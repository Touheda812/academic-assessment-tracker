```python
import pandas as pd
import matplotlib.pyplot as plt

# Load cleaned data
df = pd.read_csv('../output/cleaned_assessments.csv')

# Summary statistics
print(df.describe())

# Students needing intervention
print("Students Needing Help:\n", df[df['Intervention_Needed']])

# Average scores
avg_scores = df.groupby("Grade")[["ELA_Score", "Math_Score"]].mean()
avg_scores.plot(kind="bar", title="Average Scores by Grade")
plt.ylabel("Score")
plt.tight_layout()
plt.show()
```
